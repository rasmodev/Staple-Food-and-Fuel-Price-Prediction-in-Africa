**Suitable Imputation Approach and Method Selection**

---
**a. Why Median Imputation Is Cumbersome**

Median imputation involves filling missing values with the median of observed values within a specified group. While this method is straightforward and computationally inexpensive, it presents several limitations in this project:

- **i. Granularity of Data**: Given the complexity of the pricing data, which varies by country, administrative region, market, and time period, using a single median value might not capture the variability within these dimensions. For instance, the median price for a commodity in one country may not be representative of its price in another country with different economic conditions.

- **ii. Time Sensitivity**: Prices change over time due to inflation, seasonal effects, and market fluctuations. Median imputation does not account for temporal variations, potentially leading to misleading estimates if historical data differs significantly from current trends.

- **iii. Regional Variations**: Prices can vary significantly between regions and markets within a country. Median imputation might oversimplify these variations, leading to inaccurate imputations for specific regions or markets.

**b. Why K-Nearest Neighbors (KNN) Is Not Ideal for Initial Exploration** 

KNN imputation is a more sophisticated method that leverages the similarity between records to predict missing values. However, it requires several preprocessing steps that may hinder initial exploratory data analysissis (EDA):

- **i. Preprocessing Complexity**: KNN requires encoding categorical variables and scaling numerical features. These preprocessing steps are crucial for ensuring the KNN algorithm performs effectively but can obscure initial insights during EDA, which should focus on understanding the raw data.

- **ii. Data Splitting**: KNN involves splitting the data into features (X) and target values (y). This process, along with cross-validation for parameter tuning, adds complexity and may delay the exploratory phase, which is crucial for identifying patterns and outliers in the complete dataset.

---

**c. Suitable Imputation Approach** 
I will use**Linear Interpolation** to handle missing values. Linear interpolation is a method used to estimate missing values by connecting two known points with a straight line. When a data point is missing, the method calculates its value based on a linear progression between the points immediately before and after the missing point. This method assumes that the rate of change between two known points is constant and uses that rate to estimate the missing value.

---
**Why Linear Interpolation Works for This Project:**

**i. Time Sensitivity**:
   - Since the dataset includes price fluctuations over time (`period_date`, `year`, `month`), linear interpolation respects the temporal sequence. It estimates prices smoothly across dates while capturing seasonal and market trends.
  
**ii. Granularity of Data**:
   - Linear interpolation retains the nuances of price variations between different countries, regions, and markets by imputing values based on trends within each group, avoiding the oversimplification of using a single summary value like the median.

**iii. Handling Missing Temporal Data**:
   - Linear interpolation works well for time-series data, where missing values are often adjacent to known values in a sequence. This makes it suitable for estimating prices where values might be missing for a certain time period but available before and after.

## 2.3.3.Hypothesis Testing and Answering Key Analytical Questions

### a. Hypothesis Testing

**Hypothesis:** Influence of Region on Price
   - **Null Hypothesis (H₀):** There is no significant difference in prices based on the administrative regions or countries.
   
   - **Alternative Hypothesis (H₁):** There is a significant difference in prices based on the administrative regions or countries.

---

**Objective:** To determine if there are significant differences in prices based on the administrative regions or countries.

The **ANOVA (Analysis of Variance) Test** is best-suited for this hypothesis test for various reasons:

**i. Comparing Multiple Groups:**
   - **Country:** There are multiple countries in the dataset.
   - **Administrative Region:** There are multiple administrative regions in the dataset.
   - ANOVA is suitable for comparing the means of more than two groups, helping us understand if at least one group mean is significantly different from others.

**ii. Quantitative Dependent Variable:**
   - The dependent variable, `value` (price), is quantitative.
   - ANOVA is designed to compare the means of a quantitative dependent variable across different groups (categorical independent variables).

**iii. Categorical Independent Variables:**
   - **Country** and **Administrative Region** are categorical variables.
   - ANOVA tests how these categorical variables influence the mean of the quantitative variable.

**iv. Testing Overall Differences:**
   - ANOVA provides a method to test for overall differences among group means rather than just comparing pairs of groups.
   - This helps in identifying whether the observed differences in means are statistically significant when considering all groups simultaneously.

ANOVA is the appropriate test for determining if there are significant differences in prices based on the administrative regions or countries in the dataset.

Based on the analysis of the unique values, this is the encoding strategies:

- **i. One-Hot Encoding:** country, product_category, price_category, currency

One-hot encoding is a method where each unique value in a categorical column is converted into a new binary column. For example, the country column with 13 unique values will be transformed into 13 binary columns, each representing one of the countries. This approach is effective for columns with a relatively small number of unique values (low cardinality) because it ensures that each category is represented distinctly without introducing ordinal relationships.

- **ii. Label Encoding:** product_code, product_source

Label encoding converts each unique category into a distinct integer value. For example, product_code will be converted into integer values from 0 to 136. This approach is suitable for categorical variables where the categories do not have a meaningful ordinal relationship, but the model might benefit from a simple numerical representation.

- **Frequency Encoding:** market, administrative_region, measurement_unit, product

Frequency encoding involves replacing each category with its frequency of occurrence in the dataset. For example, a market with 184 unique values will be encoded with the count of each market's occurrences. This method is useful for high cardinality features as it transforms the categories into numerical values based on their occurrence, potentially capturing the importance of each category based on its frequency.

```python
# One-Hot Encoding
eda_df = pd.get_dummies(eda_df, columns=['country', 'product_category', 'price_category', 'currency'])

# Label Encoding
label_encoders = {}
for col in ['product_source']:
    le = LabelEncoder()
    eda_df[col] = le.fit_transform(eda_df[col])
    label_encoders[col] = le

# Frequency Encoding Function
def frequency_encoding(df, column):
    freq = df[column].value_counts()
    df[column + '_freq'] = df[column].map(freq)
    df.drop(column, axis=1, inplace=True)
    return df

# Apply Frequency Encoding to specified columns
for col in ['market', 'administrative_region', 'measurement_unit', 'product']:
    eda_df = frequency_encoding(eda_df, col)

# Display the head of the final encoded DataFrame
df_encoded = eda_df
df_encoded.head()
```

```python
# Check for numerical columns
num_cols = X_train.select_dtypes(include=['number']).columns

# Analyze unique values to identify columns to scale
for col in num_cols:
    unique_values = X_train[col].unique()
    num_unique = len(unique_values)
    title = f"Column: {col}"
    print(title)
    print('='*len(title))
    print(f"Number of unique values: {num_unique}")
    print(f"Unique values: {unique_values[:10]}")
    print()
```

---

- Many features in the dataset are on different scales ('value' in monetary units). Without scaling, the features with larger ranges or units will disproportionately affect the model's performance as they will dominate the model's learning process.
- If one feature ranges from 0 to 1000 and another ranges from 0 to 1, the feature with the larger range will dominate the model's learning process. Scaling ensures that each feature contributes equally. 
- Therefore, these features need to be within the same scale.

---

- All features are scaled to a fixed range, typically to have a mean of 0 and a standard deviation of 1, which helps to make sure that no feature dominates due to its larger numerical range. 

- Let's identify the columns to scale.

---

```python
# Function to automatically select columns for scaling
def get_numerical_columns_for_scaling(df):
    # Initialize a list to hold columns to scale
    columns_to_scale = []
    
    # Loop through each column
    for column in df.columns:
        # Get unique values for the column
        unique_values = df[column].unique()
        num_unique_values = len(unique_values)
        
        # Add the column to the list if it has more than 2 unique values
        if num_unique_values > 2:
            columns_to_scale.append(column)
    
    return columns_to_scale

# Identify numerical columns to scale
numerical_columns_to_scale = get_numerical_columns_for_scaling(X_train)
print(numerical_columns_to_scale)
```

Binary columns (0 or 1) are already normalized. Applying standard scaling to these columns might not be meaningful and could even distort their representation. That is why I am selecting features with more than 2 unique values for scaling.

```python
# Initialize the scaler
scaler = StandardScaler()

# Copy the original data to ensure the original DataFrames are not modified
X_train_final = X_train.copy()
X_eval_final = X_eval.copy()

# Fit and transform the numerical columns in the training set
X_train_final[numerical_columns_to_scale] = scaler.fit_transform(X_train_final[numerical_columns_to_scale])

# Transform the numerical columns in the testing set
X_eval_final[numerical_columns_to_scale] = scaler.transform(X_eval_final[numerical_columns_to_scale])
```

### ** Evaluation Metrics Explanation:**
i. **RMSLE (Root Mean Squared Logarithmic Error):**
   - RMSLE penalizes underestimation more than overestimation, making it suitable for scenarios where the target variable (prices in this case) spans a large range.
   - Lower values are better; the closer the RMSLE is to 0, the more accurate the model's predictions are.
   
ii. **RMSE (Root Mean Squared Error):**
   - RMSE provides an average deviation of the predicted prices from the actual prices. It is sensitive to large errors due to the squaring of differences.
   - Lower RMSE indicates better model performance.
   
iii. **MSE (Mean Squared Error):**
   - MSE gives the squared average error between predictions and actual prices, heavily penalizing larger errors. It’s closely related to RMSE.
   
iv. **MAE (Mean Absolute Error):**
   - MAE gives the average absolute difference between predicted and actual prices. It’s more intuitive and less sensitive to large errors than RMSE/MSE.

## 2.4.4. Creating a Pre-Processing Pipeline For Model Deployment

```python
# Custom Transformers

class LabelEncoderTransformer(BaseEstimator, TransformerMixin):
    def __init__(self, columns):
        self.columns = columns
        self.encoders = {}
        
    def fit(self, X, y=None):
        for col in self.columns:
            le = LabelEncoder()
            le.fit(X[col])
            self.encoders[col] = le
        return self
    
    def transform(self, X):
        X = X.copy()
        for col in self.columns:
            le = self.encoders[col]
            X[col] = le.transform(X[col])
        return X

class FrequencyEncoderTransformer(BaseEstimator, TransformerMixin):
    def __init__(self, columns):
        self.columns = columns
        
    def fit(self, X, y=None):
        return self
    
    def transform(self, X):
        X = X.copy()
        for col in self.columns:
            freq = X[col].value_counts()
            X[col + '_freq'] = X[col].map(freq)
            X.drop(col, axis=1, inplace=True)
        return X

# Define categorical and numerical feature lists
categorical_features = data.select_dtypes(include=['object']).columns.tolist()
numerical_features = data.select_dtypes(include=['number']).columns.tolist()

# Define preprocessing for categorical and numerical features
preprocessor = ColumnTransformer(
    transformers=[
        ('categorical', Pipeline(steps=[
            ('one_hot_encoder', OneHotEncoder(handle_unknown='ignore')),
            ('label_encoder', LabelEncoderTransformer(columns=['product_code', 'product_source'])),
            ('frequency_encoder', FrequencyEncoderTransformer(columns=['market', 'administrative_region', 'measurement_unit', 'product']))
        ]), categorical_features),
        ('numerical', Pipeline(steps=[
            ('scaler', StandardScaler())
        ]), numerical_features)
    ],
)

# Define the full pipeline with preprocessing and classifier
pipe = Pipeline(steps=[
    ('preprocessor', preprocessor),
    ('regressor', RandomForestRegressor(n_estimators=100, random_state=42))
])


pipe
```
